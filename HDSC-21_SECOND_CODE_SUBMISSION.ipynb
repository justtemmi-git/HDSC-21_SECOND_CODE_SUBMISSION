```python
app_df= pd.read_csv('energydata_complete.csv ', na_values={' ' : np.nan})
```

```python
app_df= app_df.drop(columns= ['date', 'lights'])
```

```python
#firstly we normalise our dataset to a common scale using the min max scaler
scaler= MinMaxScaler()
normalised_df= pd.DataFrame(scaler.fit_transform(app_df), columns= app_df.columns)
```

```python
app_feat_df= normalised_df.drop(columns='Appliances' )
app_target= normalised_df['Appliances']
```

```python
app_target.head(10)
```

```python
app_feat_df.head(10)
```

```python
#now, we split our dataset into training and testing. we have segmented the features and target variables.
x_train, x_test, y_train, y_test = train_test_split(app_feat_df, app_target, test_size= 0.3, random_state= 42)
```

```python
l_model= LinearRegression()
#fit the model to the training dataset
l_model.fit(x_train, y_train)
#obtain predictions
predicted_values= l_model.predict(x_test)
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
mae= mean_absolute_error(y_test, predicted_values)
round(mae, 3)
```

```python
#checking the RSS Residual sum of squares
rss= np.sum(np.square(y_test - predicted_values))
round(rss, 3)
```

```python
#Root Mean Square Error RMSE
from sklearn.metrics import mean_squared_error
rmse= np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
#R-squared
from sklearn.metrics import r2_score
r2_score= r2_score(y_test, predicted_values)
round(r2_score,3)
```

```python
#Ridge L2 regression
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha= 0.5)
ridge_reg.fit(x_train, y_train)
```

```python
rmse= np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
#Lasso L1 regularisation
from sklearn.linear_model import Lasso
lasso_reg= Lasso(alpha= 0.001)
lasso_reg.fit(x_train, y_train)
```

```python
rmse= np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
#comparing the results of th regularizations
def get_weights_df(model, feat, col_name):
    #this function returns the weights of every feature
    weights= pd.Series(model.coef_, feat.columns).sort_values()
    weights_df= pd.DataFrame(weights).reset_index()
    weights_df.columns= ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
l_model_weights= get_weights_df(l_model, x_train, 'Linear_Model_Weight')
ridge_weights_df= get_weights_df(ridge_reg, x_train, 'Ridge_Weight')
lasso_weights_df= get_weights_df(lasso_reg, x_train, 'Lasso_Weight')
```

```python
final_weights= pd.merge(l_model_weights, ridge_weights_df, on= 'Features')
final_weights= pd.merge(final_weights, lasso_weights_df, on= 'Features')
```
